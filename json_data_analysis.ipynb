## 3.  Data Exploration

```python
print('── Shape ──────────────────────────')
print(f'Rows: {df.shape[0]}, Columns: {df.shape[1]}')

print('\n── Column Data Types ──────────────')
print(df.dtypes)

print('\n── Missing Values ─────────────────')
print(df.isnull().sum())
```

```python
# Statistical summary of numeric columns
df.describe().round(2)
```

## 4.  Data Cleaning & Feature Engineering

```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract time features
df['month']    = df['date'].dt.month
df['month_name'] = df['date'].dt.strftime('%b')
df['quarter']  = df['date'].dt.quarter

# Compute revenue
df['revenue']  = df['quantity'] * df['unit_price']

# Categorise order size
df['order_size'] = pd.cut(
    df['revenue'],
    bins=[0, 5000, 30000, float('inf')],
    labels=['Small', 'Medium', 'Large']
)

print('✅ Feature engineering done!')
df[['date','month_name','quarter','revenue','order_size']].head(8)
```

## 5.  Aggregations & Group Analysis

```python
# Revenue by Category
cat_rev = df.groupby('category')['revenue'].sum().sort_values(ascending=False)
print('── Revenue by Category ─────────────────')
print(cat_rev.apply(lambda x: f'₹{x:,.0f}'))
```

```python
# Revenue by Region
region_rev = df.groupby('region')['revenue'].sum().sort_values(ascending=False)
print('── Revenue by Region ───────────────────')
print(region_rev.apply(lambda x: f'₹{x:,.0f}'))
```

```python
# Monthly revenue trend
monthly = df.groupby(['month','month_name'])['revenue'].sum().reset_index()
monthly = monthly.sort_values('month')
print('── Monthly Revenue ─────────────────────')
monthly[['month_name','revenue']]
```

```python
# Pivot: Category × Region
pivot = df.pivot_table(values='revenue', index='category', columns='region', aggfunc='sum', fill_value=0)
print('── Category × Region Revenue Matrix ────')
pivot
```

## 6.  Visualisations

### 6a. Revenue by Category — Bar Chart

```python
fig, ax = plt.subplots(figsize=(9,5))
colors = sns.color_palette('husl', len(cat_rev))
bars = ax.bar(cat_rev.index, cat_rev.values, color=colors, width=0.55, edgecolor='white', linewidth=1.5)
for bar, val in zip(bars, cat_rev.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1500,
            f'₹{val/1000:.0f}K', ha='center', va='bottom', fontsize=10, fontweight='bold')
ax.set_title('Total Revenue by Category', fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel('Category', fontsize=12)
ax.set_ylabel('Revenue (₹)', fontsize=12)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'₹{x/1000:.0f}K'))
plt.tight_layout()
plt.savefig('revenue_by_category.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 6b. Monthly Revenue Trend — Line Chart

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(monthly['month_name'], monthly['revenue'], marker='o', linewidth=2.5,
        markersize=8, color='#2196F3', markerfacecolor='white', markeredgewidth=2.5)
ax.fill_between(monthly['month_name'], monthly['revenue'], alpha=0.12, color='#2196F3')
for _, row in monthly.iterrows():
    ax.annotate(f"₹{row['revenue']/1000:.0f}K",
                xy=(row['month_name'], row['revenue']),
                xytext=(0, 12), textcoords='offset points',
                ha='center', fontsize=9, color='#1565C0')
ax.set_title('Monthly Revenue Trend (2024)', fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Revenue (₹)', fontsize=12)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'₹{x/1000:.0f}K'))
plt.tight_layout()
plt.savefig('monthly_trend.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 6c. Revenue Share — Pie / Donut Chart

```python
fig, axes = plt.subplots(1, 2, figsize=(13,5))

# Donut – Category
wedge_props = dict(width=0.55, edgecolor='white', linewidth=2)
axes[0].pie(cat_rev.values, labels=cat_rev.index, autopct='%1.1f%%',
            colors=sns.color_palette('husl', len(cat_rev)),
            wedgeprops=wedge_props, pctdistance=0.75, startangle=90)
axes[0].set_title('Revenue Share by Category', fontweight='bold')

# Donut – Region
axes[1].pie(region_rev.values, labels=region_rev.index, autopct='%1.1f%%',
            colors=sns.color_palette('Set2', len(region_rev)),
            wedgeprops=wedge_props, pctdistance=0.75, startangle=90)
axes[1].set_title('Revenue Share by Region', fontweight='bold')

plt.tight_layout()
plt.savefig('revenue_share.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 6d. Category × Region Heatmap

```python
fig, ax = plt.subplots(figsize=(10,4))
sns.heatmap(pivot, annot=True, fmt=',', cmap='YlOrRd',
            linewidths=0.5, linecolor='white',
            annot_kws={'size':10}, ax=ax)
ax.set_title('Revenue Heatmap: Category × Region', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Region', fontsize=11)
ax.set_ylabel('Category', fontsize=11)
plt.tight_layout()
plt.savefig('heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 6e. Quantity vs Revenue Scatter Plot

```python
palette = {'Electronics':'#1565C0','Clothing':'#E91E63','Books':'#4CAF50'}
fig, ax = plt.subplots(figsize=(10,5))
for cat, grp in df.groupby('category'):
    ax.scatter(grp['quantity'], grp['revenue'],
               label=cat, color=palette.get(cat,'grey'),
               s=120, alpha=0.85, edgecolors='white', linewidths=1.2)
for _, row in df.iterrows():
    ax.annotate(row['product'], (row['quantity'], row['revenue']),
                textcoords='offset points', xytext=(5,4),
                fontsize=7.5, alpha=0.75)
ax.set_title('Quantity vs Revenue by Category', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Quantity Sold', fontsize=11)
ax.set_ylabel('Revenue (₹)', fontsize=11)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'₹{x/1000:.0f}K'))
ax.legend(title='Category', framealpha=0.8)
plt.tight_layout()
plt.savefig('scatter.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 6f. Payment Method Distribution — Plotly Interactive

```python
pay = df.groupby('payment')['revenue'].sum().reset_index()
fig = px.bar(pay, x='payment', y='revenue', color='payment',
             color_discrete_sequence=px.colors.qualitative.Bold,
             title='Revenue by Payment Method',
             labels={'revenue':'Revenue (₹)', 'payment':'Payment Method'},
             text_auto='.2s')
fig.update_layout(showlegend=False, plot_bgcolor='white',
                  title_font_size=16, title_x=0.05)
fig.show()
```

### 6g. Quarterly Revenue — Grouped Bar (Plotly)

```python
qtr = df.groupby(['quarter','category'])['revenue'].sum().reset_index()
fig = px.bar(qtr, x='quarter', y='revenue', color='category',
             barmode='group', title='Quarterly Revenue by Category',
             labels={'revenue':'Revenue (₹)','quarter':'Quarter'},
             color_discrete_sequence=px.colors.qualitative.Set2)
fig.update_layout(plot_bgcolor='white', title_font_size=16, title_x=0.05,
                  xaxis=dict(tickprefix='Q', tickmode='linear'))
fig.show()
```

## 7. 🔗 Correlation Analysis

```python
corr = df[['quantity','unit_price','revenue','month','quarter']].corr()

fig, ax = plt.subplots(figsize=(7,5))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, linecolor='white',
            annot_kws={'size':11}, ax=ax)
ax.set_title('Correlation Matrix', fontsize=14, fontweight='bold', pad=12)
plt.tight_layout()
plt.savefig('correlation.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 8.  Top Products & Customers

```python
print('── Top 5 Products by Revenue ───────────────────────────')
top_products = (df.groupby('product')['revenue']
                  .sum()
                  .sort_values(ascending=False)
                  .head(5)
                  .apply(lambda x: f'₹{x:,.0f}'))
print(top_products.to_string())

print('\n── Top 5 Customers by Revenue ──────────────────────────')
top_customers = (df.groupby('customer')['revenue']
                   .sum()
                   .sort_values(ascending=False)
                   .head(5)
                   .apply(lambda x: f'₹{x:,.0f}'))
print(top_customers.to_string())
```

```python
# Horizontal bar — Top 5 products
tp = df.groupby('product')['revenue'].sum().sort_values().tail(5)
fig, ax = plt.subplots(figsize=(9,4))
bars = ax.barh(tp.index, tp.values,
               color=sns.color_palette('magma_r', len(tp)),
               edgecolor='white', height=0.6)
for bar, val in zip(bars, tp.values):
    ax.text(val + 500, bar.get_y() + bar.get_height()/2,
            f'₹{val/1000:.0f}K', va='center', fontsize=10, fontweight='bold')
ax.set_title('Top 5 Products by Revenue', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Revenue (₹)', fontsize=11)
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f'₹{x/1000:.0f}K'))
plt.tight_layout()
plt.savefig('top_products.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 9.  KPI Summary Dashboard

```python
total_rev   = df['revenue'].sum()
total_orders = len(df)
avg_order   = df['revenue'].mean()
top_cat     = cat_rev.idxmax()
top_region  = region_rev.idxmax()
best_month  = monthly.loc[monthly['revenue'].idxmax(), 'month_name']

kpis = {
    'Total Revenue'       : f'₹{total_rev:,.0f}',
    'Total Orders'        : total_orders,
    'Avg Order Value'     : f'₹{avg_order:,.0f}',
    'Top Category'        : top_cat,
    'Top Region'          : top_region,
    'Best Month'          : best_month,
}

print('\n' + '='*45)
print('          📊  KPI SUMMARY DASHBOARD')
print('='*45)
for k, v in kpis.items():
    print(f'  {k:<22}  {v}')
print('='*45)
```
